```python
import pandas as pd

# Load the database!
hex_maze_database = pd.read_pickle('maze_configuration_database.pkl')

# Look at the database
display(hex_maze_database)
```

```python
from hex_maze_utils import plot_hex_maze

# Now lets plot an example maze
barriers = [37, 7, 39, 41, 14, 46, 20, 23, 30]
plot_hex_maze(barriers)
```

```python
# now let's get a subset of the dataframe including only mazes with certain attributes!
df = hex_maze_database 

# for example, I want only mazes with 3 choice points
filtered = df[(df['num_choice_points'] == 3)]
print(f"There are {len(filtered)} mazes that fit this criteria.")
display(filtered)

# now I want only mazes with 3 specific choice points: 21, 20 and 16
filtered = df[(df['choice_points'] == {21, 20, 16})]
print(f"There are {len(filtered)} mazes that fit this criteria.")
display(filtered)

# now I want mazes with 16 as the ONLY choice point
filtered = df[(df['choice_points'] == {16})]
print(f"There are {len(filtered)} mazes that fit this criteria.")
display(filtered)

# or mazes with 16 as a choice point, but it's ok if there are other choice points too
filtered = df[df['choice_points'].apply(lambda x: 16 in x)]
print(f"There are {len(filtered)} mazes that fit this criteria.")
display(filtered)
```

```python
# For Jose - find max path length difference

# Load the database
df = pd.read_pickle('maze_configuration_database.pkl')

max_path_length_diff = df['path_length_difference'].max()
print(f"The max path length difference is {max_path_length_diff}.")

# I only want mazes with the max path length difference and one choice point
filtered = df[(df['path_length_difference'] == max_path_length_diff) & (df['num_choice_points'] == 1)]
print(f"There are {len(filtered)} mazes with path length difference {max_path_length_diff} and exactly one choice point:")
display(filtered)

# Lets plot one to see it
barriers = {34, 39, 42, 11, 12, 16, 21, 27, 30}
plot_hex_maze(barriers)


# TODO for Jose

# want 17, 19, 21 as path lengths
# after barrier change, still 17, 19, 21 
# want 7 barrier change sequences of length 6
```
