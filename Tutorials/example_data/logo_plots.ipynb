### Plot some options for a `hex-maze-neuro` logo

```python
import sys
sys.path.append("../..")

from src.hexmaze import plot_hex_maze, get_hex_distance, get_optimal_paths

logo_maze = {37, 7, 39, 41, 14, 46, 20, 23, 30}

rat_hex = 16
start_port = 3
end_port = 1

rat_hex_path_past = get_optimal_paths(logo_maze, start_hex=start_port, target_hex=rat_hex)[0]
rat_hex_path_future = get_optimal_paths(logo_maze, start_hex=rat_hex, target_hex=1)[0]
rat_hex_path_past_dist = {h: get_hex_distance(logo_maze, start_hex=rat_hex, target_hex=h) for h in rat_hex_path_past}
rat_hex_path_future_dist = {h: get_hex_distance(logo_maze, start_hex=rat_hex, target_hex=h) for h in rat_hex_path_future}

rat_facing = rat_hex_path_future[1]

plot_hex_maze(logo_maze, show_barriers=False, show_hex_labels=False, rat=rat_hex, rat_to=rat_facing, 
              color_by=rat_hex_path_past_dist, colormap="Blues", vmin=-15)

plot_hex_maze(logo_maze, show_barriers=False, show_hex_labels=False, rat=rat_hex, rat_to=rat_facing, 
              hex_path=rat_hex_path_past)

plot_hex_maze(logo_maze, show_barriers=False, show_hex_labels=False, rat=rat_hex, rat_to=rat_facing, 
              color_by=rat_hex_path_future_dist, colormap="Blues", vmin=-5)
```
