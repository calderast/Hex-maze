## 2. Get a subset of the database including only mazes with certain attributes
Often, we use lambda expressions to compare each row to a specific criteria.



#### Here's how to filter the database in different ways to find mazes with specifc choice points.

```python
# For example, I want only mazes with 3 choice points
filtered = df[(df['num_choice_points'] == 3)]
print(f"There are {len(filtered)} mazes with exactly 3 choice points:")
display(filtered.head())

# Now I want only mazes with 3 specific choice points: 21, 20 and 16
filtered = df[(df['choice_points'] == {21, 20, 16})]
print(f"There are {len(filtered)} mazes with 21, 20 and 16 as the 3 choice points:")
display(filtered.head())

# Now I want mazes with 16 as the ONLY choice point
filtered = df[(df['choice_points'] == {16})]
print(f"There are {len(filtered)} mazes where hex 16 is the only choice point:")
display(filtered.head())

# Or mazes with 16 as a choice point, but it's ok if there are other choice points too
filtered = df[df['choice_points'].apply(lambda x: 16 in x)]
print(f"There are {len(filtered)} mazes where hex 16 is one of the choice points:")
display(filtered.head())

# Now I want a maze where at least 30% is dead ends
filtered = df[df['dead_end_pct'] >= 30]
print(f"There are {len(filtered)} mazes where >= 30% of hexes are part of dead ends:")
display(filtered.head())

# Or a maze with at least 2 dead ends of length >= 3 hexes
filtered = df[df['num_dead_ends_min_length_3'] >= 2]
print(f"There are {len(filtered)} mazes with at least 2 dead ends at least 3 hexes long:")
display(filtered.head())
```

#### Here's how to filter the database in different ways for mazes with different reward path lengths.

```python
# For example, I want only mazes with reward path lengths 17, 19, and 21 (in that specific order)
filtered = df[df['reward_path_lengths'].apply(lambda x: x == [17, 19, 21])]
print(f"There are {len(filtered)} mazes with path lengths 17, 19, 21, with length 17 between ports 1 and 2, 19 between 1 and 3, and 21 between 2 and 3:")
display(filtered.head())

# Now I want mazes with reward path lengths 17, 19, and 21 in any order
filtered = df[df['reward_path_lengths'].apply(lambda x: set([17, 19, 21]).issubset(x))]
print(f"There are {len(filtered)} mazes with path lengths 17, 19, 21 in any order:")
display(filtered.head())
```

#### Here's how to filter based on multiple criteria.

```python
# I want to find the maximum path length difference in the database
max_path_length_diff = df['path_length_difference'].max()
print(f"The maximum path length difference is {max_path_length_diff}.")

# I only want mazes with this maximum path length difference and a single choice point
filtered = df[(df['path_length_difference'] == max_path_length_diff) & (df['num_choice_points'] == 1)]
print(f"There are {len(filtered)} mazes with path length difference {max_path_length_diff} and exactly one choice point:")
display(filtered.head())

# Let's loosen the criteria a bit - I want mazes with a path length difference of at least 6 and a single choice point
filtered = df[(df['path_length_difference'] >= 6) & (df['num_choice_points'] == 1)]
print(f"There are {len(filtered)} mazes with path length difference >= 6 and exactly one choice point:")
display(filtered.head())
```
